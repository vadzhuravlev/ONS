# Project

To jest dataset z wyborów w Rosji z 2018 roku. 
Dataset składa się z 23 kolumn:
- region_name: nazwa regionu, w którym znajdował się lokal wyborczy
- tik_name: nazwa terytorialnej komisji wyborczej (często oparta na nazwie dzielnicy)
- ulk_name: nazwa lokalu wyborczego (lokalna komisja wyborcza)
- registered_voters: liczba wyborców powiązanych z tym lokalem wyborczym
- received_ballots: liczba pustych kart do głosowania otrzymana od terytorialnej komisji wyborczej
- early_voting_ballots: liczba kart do głosowania dla tych, którzy chcą głosować przed datą główną
- onsite_ballots: liczba kart do głosowania, które są podane w lokalu wyborczym
- offsite_ballots: liczba oddanych kart do głosowania dla tych, którzy chcą głosować poza tym miejscem wyborów (głosowanie w domu)
- unused_ballots: liczba kart do głosowania, które były jeszcze niewykorzystane na koniec dnia wyborów
- found_offsite_ballots: liczba kart do głosowania, które znaleziono w urnach wyborczych poza siedzibą
- found_onsite_ballots: liczba kart do głosowania, które znaleziono w urnach wyborczych w miejscu do głosowania
- invalid_ballots: liczba kart do głosowania uznanych za nieważne (więcej niż jeden wybór, obrazy itp.)
- valid_ballots: liczba kart do głosowania uznanych za ważne (tylko jeden wybór)
- lost_ballots: liczba kart do głosowania, które zostały utracone (received_ballots - (onsite_ballots + offsite_ballots + unused_ballots))
- unaccounted_ballots: liczba kart do głosowania, która nie została uwzględniona w momencie otrzymania  (-1 * lost_ballots jeżeli lost_ballots < 0)
- baburin: liczba głosów na Baburina Siergieja Nikołajewicza
- grudinin: liczba głosów dla Grudinina Pawła Nikołajewicza
- zhirinovsky: liczba głosów na Żyrinowski Władimir Wolfowicz
- putin: liczba głosów dla Putina Władimira Władimirowicza
- sobchak: liczba głosów za Sobczak Ksenią Anatolijewną
- suraykin: liczba głosów na Suraykin Maxim Alexandrovich
- titov: liczba głosów dla Tytowa Borysa Jurjewicza
- yavlinsky: liczba głosów dla Yavlinsky'ego Grigoriya Alexeevicha
